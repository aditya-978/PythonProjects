```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.naive_bayes import MultinomialNB
from sklearn import metrics
from sklearn.metrics import classification_report, roc_curve, auc
```

```python
# Load dataset from CSV file (Ensure the CSV file is in the same directory or provide the full path)
df = pd.read_csv(r'C:\Users\goura\Desktop\Python projects\experiment - 6\Book1.csv')
```

```python
# Checking the first few rows of the dataset
print("Dataset Preview:\n", df.head())
```

```python
df['labelnum'] = df.Label.map({'pos': 1, 'neg': 0})  # Convert labels to numerical values
```

```python
# Splitting the dataset into training and testing sets
X = df['Text Documents']  # Features (Text Messages)
y = df['labelnum']  # Labels (Numerical)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

```python
# Vectorizing text data using CountVectorizer (Bag-of-Words model)
vectorizer = CountVectorizer()
X_train_dtm = vectorizer.fit_transform(X_train)
X_test_dtm = vectorizer.transform(X_test)
```

```python
# Training the Naïve Bayes classifier
nb = MultinomialNB()
nb.fit(X_train_dtm, y_train)
```

```python
# Making predictions on test data
y_pred = nb.predict(X_test_dtm)
y_pred_proba = nb.predict_proba(X_test_dtm)[:, 1]  # Get probability scores for ROC
```

```python
# Evaluating the classifier performance
accuracy = metrics.accuracy_score(y_test, y_pred)
precision = metrics.precision_score(y_test, y_pred)
recall = metrics.recall_score(y_test, y_pred)
conf_matrix = metrics.confusion_matrix(y_test, y_pred)
```

```python
# Display performance metrics
print("\nModel Performance:")
print("Accuracy:", accuracy)
print("Precision:", precision)
print("Recall:", recall)
print("Confusion Matrix:\n", conf_matrix)
```

```python
# Step 7: Visualize Confusion Matrix
plt.figure(figsize=(5, 4))
sns.heatmap(conf_matrix, annot=True, fmt="d", cmap="Blues", xticklabels=["Negative", "Positive"], yticklabels=["Negative", "Positive"])
plt.xlabel("Predicted")
plt.ylabel("Actual")
plt.title("Confusion Matrix Heatmap")
plt.show()
```

```python
# Classification Report
print("\nClassification Report:\n", classification_report(y_test, y_pred))
```

```python
# Step 8: ROC Curve
fpr, tpr, _ = roc_curve(y_test, y_pred_proba)  # Compute False Positive Rate & True Positive Rate
roc_auc = auc(fpr, tpr)  # Compute AUC (Area Under Curve)
```

```python
# Plot ROC Curve
plt.figure(figsize=(6, 5))
plt.plot(fpr, tpr, color='blue', lw=2, label=f'ROC curve (AUC = {roc_auc:.2f})')
plt.plot([0, 1], [0, 1], color='gray', linestyle='--')  # Diagonal Line
plt.xlim([0.0, 1.0])
plt.ylim([0.0, 1.05])
plt.xlabel("False Positive Rate")
plt.ylabel("True Positive Rate")
plt.title("ROC Curve")
plt.legend(loc="lower right")
plt.show()
```
